```python
import math
def freespace_loss(distance, frequency):
    #frequency is frequency(Hz)
    #distance is meters(m)
    light_speed = 300000000.0 # m/s
    return 20 * math.log10((4.0 * math.pi * distance)/(light_speed/frequency))
```

```python

#Frequency 2 at a distance of 100 meters.For 412GHz
freespace_loss(100, 2.412 * 1000000000) # => 80.0893182554 dB
```

```python
freq1 = 700 * 1000000000
freq2 = 3500 * 1000000000
```

```python
result1 = {}
for dist in range(1,8001,50):
    result1[dist] = 63 -freespace_loss(dist,freq1)
result2 = {}
for dist in range(1,8001,50):
    result2[dist] = 63 - freespace_loss(dist,freq2)
result3 = {}
for dist in range(1,8001,50):
    result3[dist] = 66 - freespace_loss(dist,freq2)
```

```python
import matplotlib.pyplot  as plt
```

```python
# giving a title to my graph
plt.title('RSRP vs Distance')

# naming the x axis
plt.xlabel('Distance (m)')
# naming the y axis
plt.ylabel('RSRP (dB)')

plt.plot(result1.keys(),result1.values(), label = "lowband")
plt.plot(result2.keys(),result2.values(), label = "highband")
plt.plot(result2.keys(),result3.values(), label = "highband with MIMO")
# specifying horizontal line type
plt.axhline(y = -110, color = 'r', linestyle = '--')
plt.legend()
plt.show()
```
